12.

- Create a 3D array and print its shape as well as ndim.

- Create a 4D array and print its shape as well as ndim. *It's hard to visualize 4-Dimensions but numpy can do it with ease!*

```python
# 2D array
a = np.array([['cat','dog', 'hamster'], ['fish','monkey','lizard']])
print(a)
print(a.shape)
print(a.ndim)
```

```python
#3D array
a1 = np.array([[['cat','dog', 'hamster'],['fish','mnokey','lizard']],[['cat','dog','hamster'],['fish','monkey','lizard']]])
print(a1)
print(a1.shape)
print(a1.ndim)
```

```python
#4D array
a2 = np.array([[['cat','dog', 'hamster'],['fish','mnokey','lizard']],[['cat','dog','hamster'],['fish','monkey','lizard']]],ndmin=4)
print(a2)
print(a2.shape)
print(a2.ndim)
```
